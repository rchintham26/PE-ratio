```python
import pandas_datareader as pdr
import pandas as pd
from collections import defaultdict
import yfinance as yf
```

```python

nasdaq_df = pd.read_csv("nasdaq_screener_1724766063447.csv")
```

```python
nasdaq_symbols = nasdaq_df['Symbol'].tolist()
filter_symbols = []

nasdaq_symbols = [symbol for symbol in nasdaq_symbols if isinstance(symbol, str)]

for symbol in nasdaq_symbols:
    if '^' not in symbol and '/' not in symbol:
        filter_symbols.append(symbol)
        
```

```python
print(filter_symbols)
```

```python
def best_pe_ratios(nasdaq_symbols, min_pe, max_pe):
    matching_stocks = defaultdict(float)
    for stock in nasdaq_symbols:
        try:
            ticker = yf.Ticker(stock)
            pe_ratio = ticker.info.get('forwardPE')
            
            if pe_ratio is not None and min_pe <= pe_ratio <= max_pe:
                matching_stocks[ticker] = pe_ratio
        except Exception as e:
            print(f"Error retrieving data for {stock}: {e}")
        
    return matching_stocks

                
```

```python
min_pe = 0.1
max_pe = 7.5

filtered_stocks = best_pe_ratios(filter_symbols,min_pe,max_pe)
print(filtered_stocks)
```

```python
d = sorted(filtered_stocks, key = filtered_stocks.get)

for key in d:
    text = d[ d]
    print(text)
    
print(d)
```
